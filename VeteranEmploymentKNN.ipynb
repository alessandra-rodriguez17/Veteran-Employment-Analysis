```python
data1['y1_emp'] = pd.to_numeric(data1['y1_emp'], errors='coerce')
data1['y5_emp'] = pd.to_numeric(data1['y5_emp'], errors='coerce')
data1['y10_emp'] = pd.to_numeric(data1['y10_emp'], errors='coerce')
```

```python
 # Group by occupation and select y1_emp to compare it with
plt.figure(figsize = (25,30))
employmentByOccupation = data1[data1['label_dod_occ_code'] != 'All occupations'].groupby('label_dod_occ_code')[['y1_emp']].mean()
sns.barplot(x = employmentByOccupation.index, y = 'y1_emp', data = employmentByOccupation)
plt.title("Military Occupations and Their Effect on Year 1 Employment")
plt.xlabel("Military Occupations")
plt.show()
```

**Using KNN on dataset from task 1 to compare military occupations versus year 1 employment rate:**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

data1['label_dod_occ_code_encoded'] = label_encoder.fit_transform(data1['label_dod_occ_code']) # encode column because it's a string

X1 = data1['label_dod_occ_code_encoded'].values.reshape(-1, 1) # make array 2D so it works as input
y2 = (data1['y1_emp'] > 0).astype(int)
X_train, X_test, y_train, y_test = train_test_split(X1, y2, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X=X_train, y=y_train)
```

```python
predicted = knn.predict(X=X_test)
expected = y_test
predicted[:20]
expected[:20]
```

```python
wrong = [(p, e) for (p, e) in zip(predicted, expected) if p != e]
wrong
```

```python
print("Model accuracy:%.2f%%" %(knn.score(X_test,y_test)*100))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

kfold = KFold(n_splits=10, random_state=11, shuffle=True)
scores = cross_val_score(estimator=knn, X=X1, y=y2, cv=kfold)

for k in range(1, 20, 2):
     knn = KNeighborsClassifier(n_neighbors=k)
     scores = cross_val_score(estimator=knn, X=X1, y=y2, cv=kfold)
     print(f'k={k:<2}; mean accuracy={scores.mean():.2%}; ' + f'standard deviation={scores.std():.2%}')
```
